#### Sum all numbers from 1 - 100 using a loop

```python
total = 0  # Initialize sum

# Loop through numbers 1 to 100 (exclusive)
for num in range(1, 101):
    total += num  # Add each number to total

# Display the result
print("The sum of numbers from 1 to 100 is:", total)
```

#### Find the largest number in a list.

```python
numbers = [23, 45, 12, 67, 89, 34]

# initialize the largest number with the first element
larger = numbers[0]

#iterate through the list
for num in numbers:
    if num > larger:
        larger = num

print(f"largest number is: {larger}")
```

#### Check if a number is a multiple of 3 or 5.

```python
num = 15

if num % 3 == 0 or num % 5 == 0:
    print(f"{num} is a multiple of 3 and 5" )
else:
    print("not a multiple")
```

#### print the multiplication table of a number

```python
number = 5

#loop through 1 - 10
for i in range(1,11):
    print(f"{number} x {i} = {number * i}")
```

#### Reverse a string using a loop.

```python
text = "Hello"

#intialize an empty string for the reverse result
reverse = ""

#iterate through the string in reverse order
for char in text:
    reverse = char + reverse

#print the reverse string
print("reverse string: ", reverse)
```

#### count vowel in a string

```python
#input
test = "python programming"

#its used to check if the input has the vowels
vowels = "aeiou"

#count the vowels from the input
count = 0

for char in test:
    if char in vowels:
        count += 1

print(count)
```

#### Check if a number is prime.

```python
#input 
number = 17

#prime numbers are greater than 1
if number <= 1:
     print("is not a prime number")
    
else:
    #assume the number is prime until proven otherwise
    is_prime = True

    #check divisibility from 2 to num-1
    for i in range(2, number):
         
         #if number is divisible by i, it's not a prime number
         if number % i == 0:
              is_prime = False
              break
    
    if is_prime:
         print(f"{number} is a prime number")
    
    else:
         print(f"{number} is not a prime number")
```
