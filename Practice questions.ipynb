#### Check if a number is prime.

```python
def is_prime(n):

 # check if the number is greater than 1
 if n <= 1:
    return False

 # If n is divisible by any number in this range, it's not prime
 for i in range(2, int(n**0.5) + 1):
  if n % i == 0:  # found a divisor
    return False  
  # if no divisors found, n is a prime
 return True

# input
number= 5

# call the function and display
if is_prime(number):
  print(f"{number} is a prime number")
else:
  print(f"{number} is not a prime number")
```

#### Print the Fibonacci sequence up to the 10th term.

```python
def sequence(n):
    fibonacci = []
    a,b = 0,1 # first two terms

    for fib in range(n):
        fibonacci.append(a) # add current term to the list
        a, b = b, a + b # update terms

    return fibonacci
    
terms = 10 # number of terms

#get the sequence
fib_seq = sequence(terms)

#print the sequence
print(fib_seq)

```

#### Remove duplicates from a list.

```python
def remove_duplicates(lst):
    
    unique_list= []

    for items in lst:
        if items not in unique_list: # only add if not already in the list
            unique_list.append(items)
    return unique_list

# Input
numbers = [1, 2, 2, 3, 4, 4, 5, 7, 7, 7]

result = remove_duplicates(numbers)
print(result)
```

#### Calculate the factorial of a number using a loop.

```python
def factorial(n):

  result = 1 # start with 1  because multiplying by 0 would give 0

  # lopp from 1 - n (inclusive)
  for i in range(1, n + 1):
    result *= i # multiply result by the current number
  return result  

# input
number = 5
print(factorial(number))
```

#### Find the sum of digits in a number.

```python
def sum_num(n):

    total = 0

    while n > 0:
        total += n % 10 # get the last digit and add to total
        n //= 10 #remove the last digit

    return total

number = 123

print(f"the sum of {number} is",sum_num(number))
```

#### Print all permutations of a 3-letter word.

```python
def permutations(word, prefix=""):
 
    """
    Recursively print all permutations of the given word.
    :param word: Remaining letters to permute
    :param prefix: Current built string
    """
    if len(word) == 0:
        print(prefix) #base case: no letters left, print the permutation
    
    else:
        for i in range(len(word)):
            #choose the current letter and premute the rest
            remaining = word[:i] + word[i + 1:] # these will remove the letter at position i from the word 
            permutations(remaining, prefix + word[i])

permutations("cat") # works with any length word
```

#### Count how many times a substring appears in a string.

```python
def count_substring(text, sub):

    count = 0

    for i in range(len(text) - len(sub) + 1):
        #check if the sunstring matches starting at position i
        if text[i:i+len(sub)] == sub:
            count += 1
    return count

# get user inputs
main_str = input("Enter a word: ")
sub_str = input("enter the substring to look for: ")

# count occurences
result = count_substring(main_str, sub_str)

print(f"{sub_str} appears {result} times in {main_str}.")
```

#### Check if a string is a palindrome.

```python
def is_palindromwe(text):

    """
    Check if the given string is a palindrome.
    A palindrome reads the same forwards and backwards.
    
    """

    #normalize the string: remove spaces make lowercase
    wording = ' '.join(text.split()).lower()

    #compare the string with itsd reverse
    return wording == wording[::-1]

word_input = input("enter a word: ")

if is_palindromwe(word_input):
    print('its a palindrome!')

else:
    print("its not a palindrome :(")
```

#### Implement bubble sort on a list.

```python
def bubble_sort(arr):

    #loop through the list multiple times
    for i in range(len(arr)):
        for j in range(0, len(arr) - i - 1):
            #swap if the current element is greater than the next
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Input
numbers = [64, 34, 25, 12, 22, 11, 90]

bubble_sort(numbers) # sort the list
print(numbers)
```

#### Print a diamond pattern of asterisks with 5 rows.

```python
def diamond(rows):
    #upper half of the diamond
    for i in range(1, rows + 1):
        print(" " * (rows - i) + "*" *(2 * i - 1))

    #lower half of the diamond
    for i in range(rows - 1, 0, -1):
        print(" " * (rows - i) + "*" *(2 * i - 1))

rows = 5
print(diamond(rows))
```

#### Calculate the Collatz sequence.

```python
def collatz(n):
    #loop until the number becomes 1
    while n != 1:
        print(n, end=" ") #print the current number on the same line

        if n % 2 == 0:
            n //= 2   # if even, divide by 2
        
        else:
            n = 3 * n + 1 # if odd, divide by 3 and add 1
    print(1) # print the last number

number = 10

result = collatz(number)
print(result)
```

#### Validate a password.

```python
def valid_pass(password):
    """
    Validates a password based on:
    - At least 8 characters
    - At least one uppercase letter
    - At least one lowercase letter
    - At least one digit
    - At least one special character
    """
    if len(password) < 8: # check the length of the password
        return False
    
    # check if the password has these characters
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False
    specials = "!@#$%^&*()-_+=[]{};:'\",.<>?/\\|`~"

    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
        elif ch in specials:
            has_special = True
    return has_upper and has_lower and has_digit and has_special

# Input
password_valid = "P@ssw0rd"

if valid_pass(password_valid):
  print(f"{password_valid} is a valid password")
    
else:
     print(f"{password_valid} is not valid password")


```

#### Simulate a guessing game.

```python
import random

def guessing_game():
    """Number guessing game without try/except."""
    print("Welcome to the Guessing Game!")
    print("Think of a number between 1 and 100.")

    secret_number = random.randint(1, 100)
    attempts = 0

    while True:
        guess = input("Enter your guess (or 'q' to quit): ").strip()

        # Quit option
        if guess.lower() == 'q':
            print(f"Game over! The number was {secret_number}.")
            break

        # Validate numeric input
        if not guess.isdigit():
            print("Please enter a valid number.")
            continue

        guess = int(guess)

        # Validate range
        if guess < 1 or guess > 100:
            print("Please enter a number between 1 and 100.")
            continue

        attempts += 1

        if guess < secret_number:
            print("Too low! Try again.")
        elif guess > secret_number:
            print("Too high! Try again.")
        else:
            print(f"Congratulations! You guessed it in {attempts} attempts.")
            break

if __name__ == "__main__":
    guessing_game()
```

#### Find the longest word in a sentence.

```python
# Function to find the longest word in a sentence
def longest_word(sentence):
    # Remove punctuation manually (basic handling)
    cleaned_sentence = ""
    for char in sentence:
        if char.isalpha() or char.isspace():  # Keep only letters and spaces
            cleaned_sentence += char

    # Split the cleaned sentence into words
    words = cleaned_sentence.split()

    # If there are no words, return None
    if not words:
        return None

    # Assume the first word is the longest initially
    longest = words[0]

    # Loop through each word to find the longest
    for word in words:
        if len(word) > len(longest):
            longest = word

    return longest

# Main program
sentence = "The quick brown fox jumps over the lazy dog"

if sentence:  # Check if input is not empty
    result = longest_word(sentence)
    if result:
        print(f"The longest word is: '{result}' with length {len(result)}")
    else:
        print("No valid words found.")
else:
    print("Please enter a non-empty sentence.")
```

#### Implement a login attempt counter.

```python
# Simple password-only login with attempt counter

MAX_ATTEMPTS = 3
CORRECT_PASSWORD = "wrongpass"

def password_login():
    attempts = 0
    while attempts < MAX_ATTEMPTS:
        password = input("Enter password: ")

        if password == CORRECT_PASSWORD:
            print("Access granted!")
            return True
        else:
            attempts += 1
            print(f"Wrong password. Attempts left: {MAX_ATTEMPTS - attempts}")

    print("Too many failed attempts. Account Locked.")
    return False

# Run the function
password_login()
```
