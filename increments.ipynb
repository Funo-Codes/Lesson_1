# Probability

## Apply Bayes' Theorem

* Conditional probability
* Diagnostic testing
* Classification
* Machine learning

```python
# this means the probability of B happening if A has happened is 90%
P_B_given_A = 0.9

P_A = 0.5 # 50% probability of A event occuring
P_B = 0.6 # 60% probability of B event occuring 

P_A_given_B = (P_B_given_A * P_A) / P_B
print(P_A_given_B)
```
